##1. 원 핫 인코딩(one-hot encoding)
-  선택해야 하는 클래스의 개수만큼 차원을 가지면서 각 클래스의 인덱스에 해당하는 원소에는 1, 나머지 원소에는 0의 값을 가지도록 하는 표현하는 방법
- 원 핫 인코딩으로 표현된 벡터를 원 핫 벡터(one-hot vector)라고 함

```python
# class : apple, banana, orange, melon
# 구분 :    0       1       2      3
'''
원 핫 벡터
  apple = [1, 0, 0, 0]
  banana = [0, 1, 0, 0]
  orange = [0, 0, 1, 0]
  melon = [0, 0, 0, 1]

구분의 수치화 때문에 잘못된 관계성 발생
  실제값 : orange
  예측값 : banana

  오차 : (2-1)^2 = 1 -> 업데이트 해야함을 앎  => 2는 오렌지를 의미하고 1은 바나나를 의미하는 것이기 때문에 아무런 관계성이 없음. 관계식 자체가 성립이 안됌

  실제값 : melon
  예측값 : apple

  오차 : (3-0)^9 = 9 -> 업데이트 해야함을 앎  => 3는 메론를 의미하고 0은 사과를 의미하는 것이기 때문에 아무런 관계성이 없음. 관계식 자체가 성립이 안됌

  ((0, 0, 1, 0) - (0, 1, 0, 0))^2 = (-1)^2 + (1)^2 = 2
  ((0, 0, 0, 1) - (1, 0, 0, 0))^2 = (-1)^2 + (1)^2 = 2
  같은 오차를 만들어냄 => 가중치의 업데이트(변화)를 동일하게 해줌

'''
```
