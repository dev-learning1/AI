###1-1. Tensor
- 텐서(tensor)는 배열(array)이나 행렬(matrix)과 매우 유사한 특수 자료구조
- Pytorch에서는 텐서를 사용하여 모델의 입력(input)과 출력(output) 그리고 모델의 매개변수들을 부호화(encode)함

```python
data = [[1, 2], [3, 4]]
x_data = torch.tensor(data)
print(x_data)
```

```python
import numpy as np
```

```python
np_array = np.array(data)
x_np_1 = torch.tensor(np_array) # 카피를 만듬(새로운 텐서, 메모리낭비)
print(x_np_1)
```

```python
x_np_1[0, 0] = 10
print(x_np_1)
print(np_array) # 데이터값 안바뀜
```

```python
x_np_2 = torch.as_tensor(np_array)  # 뷰를 만듬. 형식이 리스트여도 상관없음
print(x_np_2)
```

```python
x_np_2[0, 0] = 10
print(x_np_2)
print(np_array)
```

```python
x_np_3 = torch.from_numpy(np_array) # 뷰를 만듬. 무조건 numpy 형식이여야함
print(x_np_3)
```

```python
x_np_3[0, 0] = 20
print(x_np_3)
print(np_array)
```

```python
# 텐서 -> ndArray
np_again = x_np_1.numpy()
print(np_again, type(np_again))
```

```python
a = torch.ones(2, 3)
print(a)

b = torch.zeros(2, 3)
print(b)

c = torch.full((2, 3), 2)
print(c)

d = torch.empty(2, 3) # 음, 양 모든 수 랜덤
print(d)
```

###1-2. Tensor 속성

```python
tensor = torch.rand(3, 4)

print(f'Shape of tensor : {tensor.shape}')
print(f'DataType of tensor : {tensor.dtype}')
print(f'Device of tensor : {tensor.device}')
```

```python
# 속성 변경
tensor = tensor.reshape(4, 3)
tensor = tensor.int()
if torch.cuda.is_available(): # gpu 사용가능 여부
  tensor = tensor.to('cuda')  # cpu에 있는 변수를 gpu로 넘김. 런타임 유형변경에서 하드웨어 가속기 gpu로 설정 -> 런타임 다시 시작

print(f'Shape of tensor : {tensor.shape}')
print(f'DataType of tensor : {tensor.dtype}')
print(f'Device of tensor : {tensor.device}')
```

###1-3. Indexing과 Slicing

```python
a = torch.arange(1, 13).reshape(3, 4)
print(a)
```

```python
print(a[1])
print(a[0, -1])
```

```python
print(a[1:-1])
```

```python
print(a[:2, 2:])
```

###1-4. Transpose

```python
a = torch.arange(16).reshape(2, 2, 4)
print(a, a.shape)
```

```python
b = a.transpose(1, 2) # a.shape=(2,2,4)의 인덱스로 크기 설정 => (a[1], a[2])=(2,4). a[0]은 생략 => 자동 설정
print(b, b.shape)
```

```python
c = a.permute((2, 0, 1))  # permute(a[2], a[0], a[1]) => shape=(4,2,2)
print(c, c.shape)
```

###1-5. Tensor 연산

```python
x = torch.tensor([[1, 2], [3, 4]], dtype=torch.float32)
y = torch.tensor([[5, 6], [7, 8]], dtype=torch.float32)
print(x)
print(y)
```

```python
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x @ y)
print('*'*30)
print(torch.add(x,y))
print(torch.subtract(x, y))
print(torch.multiply(x, y))
print(torch.divide(x, y))
print(torch.matmul(x, y))
```

```python
# in-place 연산
print(x.add(y))
print(x)

print(x.add_(y))  # x에 결과를 다시 저장 (x += 10) => x와 x+y 결과값이 나옴
```

```python
z = torch.arange(1, 11).reshape(2, 5)
print(z)
```

```python
sum1 = torch.sum(z, axis=0)
sum2 = torch.sum(z, axis=1)
sum3 = torch.sum(z, axis=-1)
print(sum1, sum1.shape)
print(sum2, sum2.shape)
print(sum3, sum3.shape)
```

```python
a = torch.arange(24).reshape(4,6)
b = a.clone().detach()  # clone : 복사 / detach : 붙여넣기(변수에 넣어줌)
print(a, a.shape)
print(b, b.shape)
```
