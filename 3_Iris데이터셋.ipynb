```python
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
```

```python
iris
```

```python
print(iris['DESCR'])
```

- DESCR : 데이터셋의 정보(설명)
- data : feature data (학습데이터)
- feature_names : feature data의 컬럼 이름
- target : label data의 숫자 (정답)
- target_names : label data의 문자

```python
iris['target_names']  # 꽃 종류 이름
```

```python
# sepal length : 꽃받침의 길이
# sepal width : 꽃받침의 너비
# petal length : 꽃잎의 길이
# petal width : 꽃잎의 너비
data = iris['data']
data[:5]
```

```python
feature_names = iris['feature_names']
feature_names
```

```python
target = iris['target']
target[:5]
```

```python
iris['target_names']
```

```python
df_iris = pd.DataFrame(data, columns=feature_names)
```

```python
df_iris.head()
```

```python
df_iris['target'] = target
df_iris.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# drop(필드명, axis(열))
x_train, x_valid, y_train, y_valid = train_test_split(  # 학습 : 검증 = 7.5 : 2.5 비율로 나눔
    df_iris.drop('target', 1), df_iris['target'] # x 변수에 target빼고 대입, y 변수에 target만 대입
)
```

```python
x_train.shape, y_train.shape  # 학습 : 4개 문제, 1개 정답
```

```python
x_valid.shape, y_valid.shape  # 검증 : 4개 검증할 문제, 1개 검증할 정답 
```

```python
import seaborn as sns
```

```python
sns.countplot(y_train)
```

```python
# stratify(필드명): 학습, 검증 변수에 데이터를 균등하게 배분
x_train, x_valid, y_train, y_valid = train_test_split(
    df_iris.drop('target', 1), df_iris['target'], stratify=df_iris['target']
)
```

```python
x_train.shape, y_train.shape  # 학습 : 4개 문제, 1개 정답
```

```python
x_valid.shape, y_valid.shape  # 검증 : 4개 검증할 문제, 1개 검증할 정답
```

```python
sns.countplot(y_train)
```

```python
from sklearn.svm import SVC # 분류
from sklearn.metrics import accuracy_score
```

```python
svc = SVC()
svc.fit(x_train, y_train)
```

```python
y_pred = svc.predict(x_valid)
print('정답률 :', accuracy_score(y_valid, y_pred))  # 85% 이상이면 괜찮은거라고 봄
```
