```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
x_train = torch.FloatTensor([[1], [2], [3], [4]])
y_train = torch.FloatTensor([[50], [70], [90], [85]]) 
```

```python
# 단순 선형 회귀는 input=1, output=1
model =  nn.Linear(1, 1)  # weight, bias 모두 Linear에 들어있음
```

```python
print(list(model.parameters())) # w, b 임의의 값 출력
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
epoch_count = 3000

for epoch in range(epoch_count+1):
  H = model(x_train)

  cost = F.mse_loss(H, y_train)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch%100 == 0:
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch, epoch_count, cost.item()))
```

```python
var = torch.FloatTensor([[5.0]])
pred = model(var)
print('학습 후 입력이 5일 때 예측 값 :', pred)
```

```python
print(list(model.parameters()))
```

```python
# 다중 선형 회귀
```

```python
x_train = torch.FloatTensor([[2, 0, 0],
                             [2, 1, 0],
                             [2, 2, 1],
                             [3, 1, 1],
                             [4, 1, 2],
                             [4, 2, 2]])
y_train = torch.FloatTensor([[50], [60], [65], [70], [75], [85]])
```

```python
model =  nn.Linear(3, 1)  # 3 : w, 1 : b
```

```python
print(list(model.parameters()))
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=1e-5)  
```

```python
epoch_count = 100000

for epoch in range(epoch_count+1):
  H = model(x_train)

  cost = F.mse_loss(H, y_train)

  optimizer.zero_grad()
  cost.backward()
  optimizer.step()

  if epoch%100 == 0:
    print('Epoch {:4d}/{} Cost: {:.6f}'.format(epoch, epoch_count, cost.item()))
```

```python
var = torch.FloatTensor([[5, 2, 3]])
pred = model(var)
print('학습 후 입력이 5, 2, 3 일 때 :',pred)  # 에폭 : 3000 => 59.4 / 에폭 : 100000 => 105.4
```
