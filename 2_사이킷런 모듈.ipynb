###1-2. LinearSVC
- 클래스를 구분으로 하는 분류문제에서 각 클래스를 잘 구분하는 선을 그려주는 방식을 사용하는 알고리즘
- 지도학습 알고리즘을 사용하는 학습 전용 데이터와 결과 전용 데이터를 함께 가지고 있어서 사용이 가능

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score # 알고리즘 평가
```

```python
learn_data = [[0, 0], [1, 0], [0, 1], [1, 1]] # 4개 문제
learn_label = [0, 0, 0, 1]  # 각 문제에 대한 정답 (0:틀림, 1:정답)
```

```python
svc = LinearSVC()
svc.fit(learn_data, learn_label) # 기계 학습
```

```python
test_data = [[0, 0], [1, 0], [0, 1], [1, 1]]
test_label = svc.predict(test_data) # 정답 예측
```

```python
print(test_data,'의 예측 결과 :',test_label)
```

```python
print('정답률 :',accuracy_score([0, 0, 0, 1], test_label))
```
