##2. 데이터 전처리

```python
# train / valid 세트로 나눔
# feature와 label을 정의 (문제와 정답)
feature = ['Pclass', 'Sex', 'Age', 'Fare']
label = ['Survived']
```

```python
train[feature].head() # 문제
```

```python
train[label].head() # 정답
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_valid, y_train, y_valid = train_test_split(
    train[feature], train[label], test_size=0.2, random_state=10
)
```

- test_size : validation set에 할당할 비율(0.2 => 20%(검증 데이터 비율))
- random_state : 랜덤 시드값(아무값이나 부여)\
(suffle 한번만 실행 => train_test_split 여러 번 실행해도 처음 실행한 데이터 값으로 고정)

##3. 결측치 처리

```python
train.info()  # 결측치: Age, Cabin, Embarked
```

```python
train.isnull().sum()
```

```python
# 나이 결측치를 0으로 처리
train['Age'].fillna(0).describe() # 데이터 수정은 안함
```

```python
# 나이 결측치를 평균나이로 처리
train['Age'].fillna(train['Age'].mean()).describe()
```

###3-1. Impute
- 2개 이상의 column을 한번에 처리할 때 사용

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer(strategy='mean')  # 여러 컬럼의 각 평균값 구하기
```

```python
imputer.fit(train[['Age', 'Pclass']]) # fit()을 통해 결측치에 대한 학습을 진행
```

```python
result = imputer.transform(train[['Age', 'Pclass']]) # 결측치를 평균값으로 변환
```

```python
result
```

```python
train[['Age', 'Pclass']] = result
```

```python
train[['Age', 'Pclass']].isnull().sum()
```

```python
train[['Age', 'Pclass']].describe()
```

```python
train['Embarked'].fillna('S') # 'Embarked'에서 'S'라는 문자가 가장 많음
```

```python
imputer = SimpleImputer(strategy='most_frequent') # 최빈값(가장 많은 문자)
```

```python
result = imputer.fit_transform(train[['Embarked', 'Cabin']])
```

```python
train[['Embarked', 'Cabin']] = result
```

```python
train[['Embarked', 'Cabin']].isnull().sum()
```

##5. One Hot Encoding
- 잘못된 관계를 학습할 수 있음
- 그것을 방지하고자 카테고리컬하게 해줌

```python
train = pd.read_csv('https://bit.ly/fc-ml-titanic')
```

```python
train['Embarked'].value_counts()
```

```python
train.info()
```

```python
# NaN값이 포함되어 있으면 LabelEncoder가 정상적으로 동작하지 않음
train['Embarked'] = train['Embarked'].fillna('S')
```

```python
train['Embarked'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder

train['Embarked_num'] = LabelEncoder().fit_transform(train['Embarked'])
```

```python
train['Embarked'].value_counts()
```

```python
# 'S'=2, 'Q'=1 관계성을 Q+Q = S 라고 학습할 수 있음
# 따라서 독립적인 데이터는 별도의 column으로 분리하고 각각의 컬럼에 해당 값에만 True, 나머지 False값을 갖게 하는 것
pd.get_dummies(train['Embarked'])[:6]
```
